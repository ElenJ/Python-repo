```python
# impute missing values
# replace missing in column "Sex" by taking the mode of the data:
#df['sex'].fillna(df['sex'].mode()[0], inplace=True)
df['sex'] = df['sex'].fillna(df['sex'].mode()[0])
df.isnull().sum(axis=0)
```

```python
display(df)
```

```python
#bill_length_mm	2
##bill_depth_mm	2
#flipper_length_mm	2
###body_mass_g	2
#sex	0
#year	0

df['bill_length_mm'] = df['bill_length_mm'].fillna(df['bill_length_mm'].median)
df['bill_depth_mm'] = df['bill_depth_mm'].fillna(df['bill_depth_mm'].median)
df['flipper_length_mm'] = df['flipper_length_mm'].fillna(df['flipper_length_mm'].median)
df['body_mass_g'] = df['body_mass_g'].fillna(df['body_mass_g'].median)
df.isnull().sum()
```

```python
df.dtypes
```

```python
#convert multiple object columns to categoric
df['sex'] = pd.Categorical(df['sex'])
df['species'] = pd.Categorical(df['species'])
df['island'] = pd.Categorical(df['island'])
# Identify object columns
object_cols = df.select_dtypes(include='object').columns

# Convert object columns to numeric, coercing errors to NaN
for col in object_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check the data types after conversion
print(df.dtypes)

# You might want to check for new missing values introduced by the coercion
print(df.isnull().sum())
```

```python
# Call the .median() method to get the median value, then pass that value to fillna
df['bill_length_mm'] = df['bill_length_mm'].fillna(df['bill_length_mm'].median())
df['bill_depth_mm'] = df['bill_depth_mm'].fillna(df['bill_depth_mm'].median())
df['flipper_length_mm'] = df['flipper_length_mm'].fillna(df['flipper_length_mm'].median())
df['body_mass_g'] = df['body_mass_g'].fillna(df['body_mass_g'].median())


df['sex'] = df['sex'].fillna(df['sex'].mode()[0])
df.isnull().sum()
```

```python
# group the df by species and return mean, median, n for all numeric columns:
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
#print(numeric_cols)
mycols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
summary_df = df.groupby(['species','year'])[mycols].agg(['mean', 'median', 'count']).round(1)
# Transpose the resulting DataFrame
transposed_aggregated_df = summary_df.T

# Display the transposed DataFrame
transposed_aggregated_df
```
